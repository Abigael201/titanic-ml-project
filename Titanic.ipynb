```python
def fill_in_any_values(cols): # Function is defined as fill_in_any_values
  age = cols[0]
  pclass = cols[1]

  if pd.isnull(age):
    if pclass == 1:
      return round(train_data[train_data['Pclass'] == 1]['Age'].mean())
    elif pclass == 2:
      return round(train_data[train_data['Pclass'] == 2]['Age'].mean())
    elif pclass == 3:
      return round(train_data[train_data['Pclass'] == 3]['Age'].mean())
  else:
        return age

train_data['Age'] = train_data[['Age','Pclass']].apply(fill_in_any_values,axis=1) # Changed to fill_in_any_values
```

```python
sex = pd.get_dummies(train_data['Sex'],drop_first=True)
```

```python
embarked = pd.get_dummies(train_data['Embarked'],drop_first=True)
```

```python
train_data.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
train_data = pd.concat([train_data,sex,embarked],axis=1)
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X = train_data.drop('Survived',axis=1)
y = train_data['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)

predictions = svm.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}

grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)

grid.fit(X_train,y_train)
grid_predictions = grid.predict(X_test)
```

```python
from sklearn.experimental import enable_halving_search_cv  # This line enables the experimental feature
from sklearn.model_selection import HalvingGridSearchCV # Now you can import HalvingGridSearchCV

param_grid = {'C':[0.5,1,10,50,100,1000],'gamma':[1,0.1,0.001,0.0001,0.00001,0.000001]}

# If you intended to use GridSearchCV, keep the following line:
# grid = GridSearchCV(SVC(),param_grid,refit=True)

# If you intended to use HalvingGridSearchCV, change GridSearchCV to HalvingGridSearchCV:
grid = HalvingGridSearchCV(SVC(),param_grid,refit=True) # Changed to HalvingGridSearchCV

grid.fit(X_train,y_train)
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test,grid_predictions ))
print(confusion_matrix(y_test,grid_predictions ))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train,y_train)

lr_predictions = lr.predict(X_test)

print(classification_report(y_test,lr_predictions ))
print(confusion_matrix(y_test,lr_predictions ))
```

```python
from sklearn.neighbors import KNeighborsClassifier

error_list = []

for i in range(1,40):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  knn_predictions = knn.predict(X_test)
  error_list.append(np.mean(knn_predictions != y_test))

```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Load the training and test data
# Check if the files are in the current directory or provide the full path if they are located elsewhere
try:
    train_data = pd.read_csv('input/train.csv')
    test_data = pd.read_csv('input/test.csv')
except FileNotFoundError:
    # If files are not found in 'input' folder, try loading from current directory
    train_data = pd.read_csv('train.csv')
    test_data = pd.read_csv('test.csv')


# Select features and target variable
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X_train = train_data[features].copy()
y_train = train_data['Survived']
X_test = test_data[features].copy()

# Preprocessing: Convert categorical variables to numeric
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# Align columns in the test and train data
X_train, X_test = X_train.align(X_test, join='left', axis=1)
X_test = X_test.fillna(0)

# Initialize and train the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Generate predictions
predictions = model.predict(X_test)

# Create a submission DataFrame
submission = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': predictions})

# Save the submission file
submission.to_csv('submission.csv', index=False)
```
